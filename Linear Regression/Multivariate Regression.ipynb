```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# load dataset
diabetes = load_diabetes()
X = diabetes.data
Y = diabetes.target
```

```python
# split into test and training sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
# Setting random_state ensures the dataset is split the same way every time you run, making results reproducible.
# Removing random_state will create a new random split each time.
# The number 42 has no special meaning—you can use any number for consistency.
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```
