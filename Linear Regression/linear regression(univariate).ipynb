```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
# model_selection is used for splitting data into training and test sets
```

```python
# scikit-learn comes with a few small standard datasets that do not require to download any file from some external website
# eg: load_diabetes(*[, return_X_y, as_frame, scaled])

X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

```python
X = X[:, 2] # extracting the 3rd column
X = X.reshape((-1,1)) # making the single column to 2 dimensional
# Extracting one feature is useful for simple analysis, learning, visualization, and testing models before working with multiple features.
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# this shows the general demonstration of the model
y_pred = model.predict(X_test)

# for predicting individual person data, use the following code instead of the above code
# new = np.array([[0.03]])    # make sure to be in 2d-array
# y_pred = model.predict(new)
# print(y_pred)
```

```python
plt.scatter(X_test, y_test,  color='black') # create points of test data
plt.plot(X_test, y_pred, color='blue', linewidth=3) # plotting new predicted values line on test data
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
